```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('shampoo_sales.csv')
```

```python
df.dtypes
```

```python
df['Datetime'] = pd.to_datetime(df['Month'],format='%m-%d')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Sales'].plot(kind='hist', bins=10, title="Sales", color='red')
plt.show()
```

```python
df['Sales'].plot(kind='line', figsize=(8,4), title="Sales", color='red')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Sales'], color = 'red')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title("Monthly Sales of Shampoo")
plt.show()
```

```python
df['month'] = df['Datetime'].dt.month
df['day'] = df['Datetime'].dt.day

df.head()
```

```python
df['lag_1'] = df['Sales'].shift(1)
df
```

```python
df['lag_2'] = df['Sales'].shift(2)
df
```

```python
df['rolling_mean_5'] = df['Sales'].rolling(5).mean()
df['rolling_std_5'] = df['Sales'].rolling(5).std()
df['rolling_min_5'] = df['Sales'].rolling(5).min()
df['rolling_max_5'] = df['Sales'].rolling(5).max()
df
```

```python
df['expanding_mean_5'] = df['Sales'].expanding(5).mean()
df['expanding_std_5'] = df['Sales'].expanding(5).std()
df['expanding_min_5'] = df['Sales'].expanding(5).min()
df['expanding_max_5'] = df['Sales'].expanding(5).max()
df
```
