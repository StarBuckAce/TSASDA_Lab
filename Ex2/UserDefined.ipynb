```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
values = np.random.randn(100)
```

```python
values
```

```python
dates = pd.date_range(start='2025-01-01',end='2025-04-10',freq='D')
```

```python
dates
```

```python
df = pd.DataFrame(values, index=dates, columns=['value'])
```

```python
df.head()
```

```python
df['value'].plot(kind='hist',bins =20, title='value', color='red')
ax = plt.gca()
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
plt.show()
```

```python
df['value'].plot(kind='line', figsize=(8,4), title = 'Value', color='red')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['value'], color='red')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Synthetic Time Series Dataset")
plt.show()
```

```python
df['year'] = df.index.year
df['month'] = df.index.month
df['day'] = df.index.day
df['weekday'] = df.index.weekday
df.head()
```

```python
df['lag_1']= df['value'].shift(1)
df
```

```python
df['lag_2']= df['value'].shift(2)
df
```

```python
df['rolling_mean_5'] = df['value'].rolling(5).mean()
df['rolling_std_5'] = df['value'].rolling(5).std()
df['rolling_min_5'] = df['value'].rolling(5).min()
df['rolling_max_5'] = df['value'].rolling(5).max()
df
```

```python
df['expanding_mean_5'] = df['value'].expanding().mean()
df['expanding_std_5'] = df['value'].expanding().std()
df['expanding_min_5'] = df['value'].expanding().min()
df['expanding_max_5'] = df['value'].expanding().max()
df
```
